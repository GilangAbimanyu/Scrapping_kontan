mencari link konten

```python
# =====================mencari link tiap web
# halaman = requests.get(link).text
kumpulan_link=[]
for pagilink in new_links:
    print(pagilink)
    nresponse = requests.get(pagilink)
    nsoup = BeautifulSoup(nresponse.text, 'lxml')
    # print(soup)
    articles = nsoup.select('.list-berita li')
    for article in articles:
        a = article.findAll('a')
        for link in a:
            href=link.get('href')
            comicUrl="https:"+href
            kumpulan_link.append({ 
                comicUrl
                # href
            })
```

```python

result=[]
for i in kumpulan_link:
    if i not in result:
        result.append(i)
print(result)
```

```python
#===============Kumpulan paragraf
import urllib3
kumpulan_paragraf = []

# header
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0',
    'ACCEPT' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'ACCEPT-ENCODING' : 'gzip, deflate, br',
    'ACCEPT-LANGUAGE' : 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
    'REFERER' : 'https://www.google.com/'
}

for kl in newlist_link:
    for link in kl:
        response=requests.get(link, headers)
        # halaman = requests.get(link).text
        nsoup = BeautifulSoup(response.text, 'lxml')
        # print(soup)
        articles = nsoup.find_all("div","sec_kiri")
    article_lists=[]
    for p in articles:
        title=p.find_all('h1','jdl_dtl')   
        teks=p.find_all('div',{"class":'ctn'})
        date=p.find_all("div",{"class":["tglbrt","fs13 color-gray mar-t-10"]})
        penulis=p.find_all(['span','b'],['sumber','penulis','showHide dbl-recommendation'])
        for a in title:
            titles=a.get_text()
        for b in date:
            dates=b.get_text()
        for c in penulis:
            writer=c.get_text()
        for d in teks:
            text=d.get_text()
        print(title,date,penulis,teks)
        kumpulan_paragraf.append({
            "titles":titles,
            "dates":dates,
            "writer":writer,
            "text":text
            })
        print(titles)
```
